```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {
    "Age Group": ["0 to 14 Years", "15 to 24 Years", "25 to 54 Years", "55 to 64 Years", "65+ Years"],
    "Population (in millions)": [350, 230, 580, 140, 120]
}

df = pd.DataFrame(data)

# Create bar chart
plt.figure(figsize=(10, 6))
plt.bar(df["Age Group"], df["Population (in millions)"], color=['gold', 'skyblue', 'blue', 'orange', 'pink'])
plt.title("Population Distribution by Age Group", fontsize=16)
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Population (in millions)", fontsize=12)

# Add value labels
for index, value in enumerate(df["Population (in millions)"]):
    plt.text(index, value + 5, str(value), ha='center', fontsize=10)

plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
